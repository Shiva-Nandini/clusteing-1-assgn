```python
df.head()
```

```python
df.boxplot()
```

```python
# Normalization function 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df_df = scaler.fit_transform(df.iloc[:,1:])
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_df_df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(4, random_state=40)
clusters_new.fit(scaled_df_df)
```

```python
clusters_new.labels_
```

```python
#Assign clusters to the data set
df['clusterid_new'] = clusters_new.labels_

#these are standardized values.
clusters_new.cluster_centers_
```

```python
df.groupby('clusterid_new').agg(['mean']).reset_index()
```

# Hierarchy

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import seaborn as sn
```

```python
df = pd.read_csv("C:\\Users\\DELL\\Downloads\\crime_data.csv");
```

```python
df
```

```python
df=df.rename({'Unnamed: 0':'named'},axis=1)
```

```python
df
```

```python
df=df.drop('named',axis=1)
```

```python
df
```

```python
df.info()
```

```python
df.boxplot()
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(df.iloc[:,1:])
```

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='average'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'average')
```

```python
# save clusters for chart
y_hc = hc.fit_predict(df_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
df_norm['h_clusterid']=Clusters
df_norm.sort_values("h_clusterid")
```
